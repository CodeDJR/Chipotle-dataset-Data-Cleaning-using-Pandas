# **HANDLING SPECIAL CHARACTERS IN TEXT BASED COLUMNS**

```python
import re
```

```python
def has_special_characters(s):
    # This regex matches any character that is not a word character or whitespace
    return bool(re.search(r'[^\w\s]', s))
chipotle['Has Special Characters'] = chipotle['item_name'].apply(has_special_characters)
```

```python
special_chars_df = chipotle[chipotle['Has Special Characters']]
print(special_chars_df)
```

```python
chipotle['item_name'] = chipotle['item_name'].str.replace(r'[^\w\s]', ' ', regex=True)
```

```python
chipotle['item_name'].value_counts()
```

```python
chipotle['choice_description'] = chipotle['choice_description'].str.replace(r'[^\w\s,]', '', regex=True)
```

```python
chipotle['choice_description'].value_counts()
```

```python
chipotle.head(10)
```

```python
chipotle.drop('Has Special Characters', axis =1)
```

Finally removed the speacial characters in the text based columns for consistency and data integrity.
